## Regression Analysis
---
Here we plot graphs of 4 different genres based on median sales (in millions) against the years. 
What this is showing us, is how trends of different games genres change over the years. The plot represents two things. The blue line is showing the exact means sold per year of that genre, and the black line is a regression line showing the overall trend of sales over the years. We can see that over time, most genres dont change much or even decline. For these graphs we decided to only pick 4 genres but almost all of them follow the same trend of decline in sales over the years. We dont really know what is the main cause of this slump but we do know that it is hitting all of these genres pretty equally.

To do this we use python library numpy for calculations, and python Regression analysis to create a linear regression. Then we use a seaborn plot to plot this data so it is easily read. 

```python
    genre = yearly_avg_sales['Shooter']
    Regr_Coeff = []
    Regr_MSE = []
    fig, axes = plt.subplots()
    
    xvals = np.transpose(np.matrix(yearly_avg_sales.index))
    count = 0
    yvals = genre.to_frame()
    regr = linear_model.LinearRegression()
    
    regr.fit(xvals,yvals)
    MSE = np.mean((regr.predict(xvals)-yvals)**2)
    Regr_Coeff.append(regr.coef_[0][0])
    Regr_MSE.append(MSE[0])
    genre.plot(ax=axes)
    axes.plot(xvals,regr.predict(xvals), color='black')
    
    y_lims = axes.get_ylim()
    y_loc = 0.85*(y_lims[1]-y_lims[0])+y_lims[0]
    axes.set_title('Shooter')
    axes.set_xlabel('Year')
    axes.set_ylabel('Mean')
    count+=1
    fig.tight_layout()
    
```

```python
yearly_avg_score = pd.pivot_table(data,index=['Year_of_Release'],
                     columns=['Genre'],values=['Critic_Score'])
yearly_avg_score = yearly_avg_score.dropna()
yearly_avg_score.columns = yearly_avg_score.columns.get_level_values(1)


yearly_avg_score
```

## Mean Critic Scores per Genre by Year
---
This is another chart to help us understand what causes reasons for games of different genres to sell differently over time. Here we are looking at how critics rate games of different genres over time to see if there is any correlation. 

We create this chart by using the Critic_Score field of the original graph, and find the mean of all games of that genre per year. 

```python
score_list = []
for genre in yearly_avg_score.columns:
    temp_list = yearly_avg_score[genre].values    
    sum = 0    
    for i in temp_list:
        sum += i     
    score_list.append(int(sum / 17))
score_graph = pd.DataFrame({'Genre':yearly_avg_score.columns, 'Critic Score':score_list})
score_graph
```

### Critic averages over all time per genre
Using this data we can understand that genres with higher average ratings sell better than genres with lower rating.

```python
ax = score_graph.plot.bar(x='Genre', y='Critic Score', rot=30, ylim=(65,80))
```

#### This is a bar graph represntation of Genre vs. Critic Scores over all time



```python
Mean_User = pd.pivot_table(data,index=['Year_of_Release'],
                     columns=['Genre'],values=['User_Score'])
Mean_User.columns = Mean_User.columns.get_level_values(1)
Mean_User = Mean_User.dropna()
Mean_User
```

## Mean User Scores per Genre by Year
---
This is another chart to help us understand what causes reasons for games of different genres to sell differently over time. Here we are looking at how users rate games of different genres over time to see if there is any correlation. 

We create this chart by using the User_Score field of the original graph, and find the mean of all games of that genre per year. 

```python
user_score_list = []

for genre in Mean_User.columns:
    temp_list = Mean_User[genre].values
    
    tot = 0
    
    for i in temp_list:
        tot += i
        
    user_score_list.append(int(tot / 17))
    
user_score_graph = pd.DataFrame({'Genre':Mean_User.columns, 'val':user_score_list})
user_score_graph
```

### User averages over all time per genre
Using this data we can understand that genres with higher average ratings sell better than genres with lower rating.

```python
df = pd.DataFrame({'User_Scores': user_score_list, 'Critic Scores': score_list}, index=Mean_User.columns)
ax = df.plot.bar(rot=50, ylim=(65,80))
```

#### This is a bar graph represntation that shows user scores alongside critic scores, it shows the difference between it.

When we compare the scores from User's and Critic's we see that Users have more varience in the scores that they pick while critics are usually pretty moderate with the scores they give. But this data does not help us see a link between how well people like playing certain games and the decline in gaming as a whole. The one thing we will note is that shooters are ranked pretty low by both Critics and Users which is also odd because we thought that the reason there sales were goin up was because of a amount of user approval.
